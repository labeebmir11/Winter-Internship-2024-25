```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
dataset=pd.read_csv("house prediction.csv")
```

```python
dataset
```

```python
data = dataset.copy()
```

```python
data
```

```python
data.columns
```

```python
features_with_nan=[i for i in data.columns if data[i].isnull().sum()>=1]
```

```python
features_with_nan
```

```python
#now we calculae the mean salesprice where the info is missing or present
for feature in features_with_nan:
    #create a variable indicating 1 as missing and 0 otherwise
     data[feature]=np.where(data[feature].isnull(),1,0)
    
     #abc
     data.groupby(feature)["SalePrice"].mean().plot.bar()
     plt.title(feature)
     plt.show()
```

```python
discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<=10]
```

```python
data=dataset.copy()
for feature in discrete_feature:
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title(feature)
    plt.show()
```
