```python
x=np.arange(0,20)
y = np.arange(2,22)
```

```python
type(x)
```

```python
#plotting via matplotlib
#scatter plots
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Scattered graph for x and y")
plt.savefig("scattered.png")
```

```python
#create subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
#sine cos
```

```python
np.pi
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.title("sine and cos wave form")

#plot the points using matplotlib
plt.plot(x,y_sin,'r--')
plt.plot(x,y_cos)
plt.grid(True)
```

```python
#Bar plot
x=[2,3,4,5]
y=[22,11,44,55]

plt.bar(x,y, color="y")
plt.grid(True)
plt.title("Bar Graph")
```

```python
#Bar plot
x=[2,3,4,5]
y=[22,11,44,55]
x1=[1,2,3,4]
y1=[25,17,45,49]
plt.bar(x1,y1,color='g')
plt.bar(x,y, color="r")
plt.grid(True)
plt.title("Bar Graph")
```

```python
#Pie Chart
label1="Grade1", "Grade2", "Grade3", "Grade4"
size1=[200, 130,240,200]
color1=["red", "orange", "yellow", "green"]
explode1=(0.2,0,0,0)
plt.pie(size1,labels=label1, colors=color1,shadow=True, explode=explode1)
```
